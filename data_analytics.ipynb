```python
# data_analytics.csv

# update the column names to match the names you have in your csv file for data_analytics

# create a pandas dataframe
dfa = pd.read_csv(
    '/content/data_analytics.csv',
    sep=',',
    names = ['Page title and screen class', 'Views', 'Active users', 'Views per active user', 'Average engagement time per active user', 'Event count',
             'Key events', 'Total revenue'],
    skiprows=1
)

# Display the first few rows of the DataFrame
print(dfa.head())
```

```python
# data_survey.csv

# update the column names to match the names you have in your csv file for data_survey

# create a pandas dataframe
dfs = pd.read_csv(
    '/content/data_survey.csv',
    sep=',',
    names = ["Timestamp", "What's your favorite color?", "What season do you enjoy the most?"],
    skiprows=1
)

# Display the first few rows of the DataFrame
print(dfs.head())
```

```python
# how many rows and columns are in the dataframe dfa?
# 14 rows, 8 columns

# how many rows and columns are in the dataframe dfs?
# 8 rows, 3 columns
```

```python
# Analytics

# Do one or more explorations with dfa
# no interaction is required

# make a title
plt.title('Figure 1: Page Popularity')

# filter out rows with NaN in 'Page title and screen class' and 'Views'
dfa_filtered = dfa.dropna(subset=['Page title and screen class', 'Views']).copy()

# remove the row where 'Views' column itself contains 'Views' (which is the actual header not the data)
dfa_filtered = dfa_filtered[dfa_filtered['Views'] != 'Views']

# make sure 'Views' is a number
dfa_filtered['Views'] = pd.to_numeric(dfa_filtered['Views'])

# make plot
plt.bar(dfa_filtered['Page title and screen class'], dfa_filtered['Views'])
# labels
plt.xlabel('Page Title')
plt.ylabel('Number of Views')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# show
plt.show()


# reveal everything
```

```python
import matplotlib.pyplot as plt

# Survey
# Do one or more explorations with dfs
# no interaction is required

# Making a pie chart based on Google Colab's guide here: https://colab.research.google.com/drive/11Gd3rmm7TR-ugw4blV6RRko-zf-d6fmF#scrollTo=a7voqDZonnxg
x = dfs["What's your favorite color?"].value_counts()
Labels_color = x.index

# Define colors for the pie chart to match the labels in color popularity chart
pie_colors = [color.lower() for color in Labels_color]

#now for the seasons data:
y = dfs['What season do you enjoy the most?'].value_counts()
Labels_season = y.index

# Create a blank canvas
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Favorite color pie chart
axes[0].pie(x, labels=Labels_color, autopct='%.1f%%', colors=pie_colors)
axes[0].set_title('Figure 2: Color Popularity')

# Favorite season pie chart
axes[1].pie(y, labels=Labels_season, autopct='%.1f%%')
axes[1].set_title('Figure 3: Season Popularity')

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```
